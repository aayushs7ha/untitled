```python
# Getting a list of VIN and mobile no to merge the dataset
latest_ro_mobile_no = ro_data[['VIN', 'MOBILE_NO']]
print(latest_ro_mobile_no.shape)
latest_ro_mobile_no.head(3)
```

```python
# Taking only VIN and Mobile no from Golden Repository
gr_vin_mobile = gr[['VIN', 'MOBILE_NUM']]
print(gr_vin_mobile.shape)
gr_vin_mobile.head(3)
```

```python
# Merging RO mobile and Golden Repository Mobile no
first_merged = pd.merge(latest_ro_mobile_no, gr_vin_mobile,how="left", on=["VIN"])
print(first_merged.shape)
first_merged.head()
```

```python
# Renaming the columns in the data set
first_merged.rename({'MOBILE_NO': 'RO_Mobile_No', 'MOBILE_NUM': 'GR_Mobile_no'}, axis=1, inplace=True)
```

```python
# Checking whether the columns have been renamed or not
first_merged.head(3)
```

```python
first_merged.dtypes
```

```python
# Cleaning phone numbers in the data frame = df[df['reason'].str.len() <= 10]
clean_phone_no = first_merged[first_merged['RO_Mobile_No'].str.len() == 10]
print(clean_phone_no.shape)
clean_phone_no.head(3)
```

```python
clean_phone_no1 = clean_phone_no[clean_phone_no['RO_Mobile_No'] != 'SATISH KUM']
```

```python
# Changing the dtype of RO mobile no
clean_phone_no1['RO_Mobile_No'] = clean_phone_no1.RO_Mobile_No.astype(float)
```

```python
# Adding a column as final Mobile no
clean_phone_no1['Verified_mob_no'] = np.where(clean_phone_no1['RO_Mobile_No'] == clean_phone_no1['GR_Mobile_no'], clean_phone_no1['GR_Mobile_no'],  clean_phone_no1['RO_Mobile_No'])
print(clean_phone_no1.shape)
clean_phone_no1.head(3)
```

```python
clean_phone_no1.isnull().sum()
```

```python
# Now loading HMM mobile no data
# Then replacing the verified mobile no with Connectivity mobile no

# Loading the HMM Data
hmm = pd.read_excel('HMM_Members_Phone_no.xltm')
print(hmm.shape)
hmm.head(3)
```

```python
# Renaming the columns
hmm.rename({'Phone': 'HMM_Mobile_no'}, axis = 1, inplace = True)
hmm.head(2)
```

```python
# Adding HMM contacts to verified mobile no
hmm_merged = pd.merge(clean_phone_no1, hmm,how = "left", on = ["VIN"])
print(hmm_merged.shape)
hmm_merged.head()
```

```python
hmm_merged.isnull().sum()
```

```python
32671-26415
```

Out of 32671 records in RO Data only 6256 have been regitered on HMM. So only for these 6256 records HMM phone no will be taken as Verified mobile no.

```python
# Updating the mobile no for these 12026 records
hmm_merged['Verified_mob_no'] = np.where(hmm_merged['HMM_Mobile_no'].notnull(), hmm_merged['HMM_Mobile_no'], hmm_merged['Verified_mob_no'] )
print(hmm_merged.shape)
hmm_merged.head(3)
```

```python
# Loading the connectivity data
connectivity = pd.read_csv('Connectivity_Registered_Mobile_User_data.csv')
print(connectivity.shape)
connectivity.head(3)
```

### Adding the complaint Yes/No column to the RO Data

```python
# Loading the complaint data
complaint = pd.read_excel("Complaints_data_YTD-Big_Data.xltm")
print(complaint.shape)
complaint.head(3)
```

```python
# Checking for null values
complaint.isnull().sum()
```

```python
# Adding a complaint yes/no column, for all the rows in this data frame the value will be 1
complaint['Complaint'] = 'Yes'
print(complaint.shape)
complaint.head(3)
```

```python
# Adding the complaint target variable to RO Data
final_data_with_target_variable = pd.merge(updated_ro_data, complaint[['VIN', 'Complaint']], how = 'left', left_on = 'VIN', right_on = 'VIN')
print(final_data_with_target_variable.shape)
final_data_with_target_variable.head(3)
```

```python
56835-55958
```

```python
# Removing duplicates
final_data_with_target_variable.drop_duplicates(subset = ['VIN'], keep= 'first')
print(final_data_with_target_variable.shape)
final_data_with_target_variable.head(3)
```

```python
# Filling the Complaint column missing values with 'No'
final_data_with_target_variable['Complaint'].fillna('No', inplace=True)
print(final_data_with_target_variable.shape)
final_data_with_target_variable.head(3)
```

```python
final_data_with_target_variable['Complaint'].value_counts()
```

Now warranty data, and PSF Data needs to be added in this dataframe.
In case of PSF data, add PSF taken date and then find the difference between RO Dtime and PSF taken date. 
Verify the warranty data.

```python
# Loading the warranty data
warranty = pd.read_excel('VIN_warranty_data.xltm')
print(warranty.shape)
warranty.head(3)
```

```python
# Removing duplicates from warranty data
warranty.drop_duplicates(subset = ['VIN'], keep = 'first')
print(warranty.shape)
warranty.head(3)
```

```python
# Merging this data with RO data
master_data = pd.merge(final_data_with_target_variable, warranty[['VIN','SALE_DATE', 'BASIC_WRNTY', 'EXTB_VIN', 'EXTB_WRNTY_NO', 'EXTB_APPL_STRT_DATE', 'EXTB_APPL_FNSH_DATE']], how = 'left', on = 'VIN')
print(master_data.shape)
master_data.head(3)
```

```python
58310-56835
```

```python
# Removing duplicates
master_data.drop_duplicates(subset = ['VIN'], keep = 'first')
print(master_data.shape)
master_data.head(3)
```

```python
# Loading the 3rd day PSF data
psf = pd.read_excel('3rd_Day_PSF_Sample_Data.xltm')
print(psf.shape)
psf.head(3)
```

```python
# Removing duplicates from PSF Data
psf = psf.drop_duplicates(subset = ['FLWUP_VIN'], keep = 'first')
print(psf.shape)
```

For merging the dataset remove duplicates before merging. So that the merged data set doesn't get duplicate rows.

```python
# Taking into consideration only the follow up customer status column and merging with RO Data
updated_master_data = pd.merge(master_data, psf[['FLWUP_VIN','FLWUP_STATUS']], how = 'left', left_on = 'VIN', right_on = 'FLWUP_VIN')
print(updated_master_data.shape)
updated_master_data.head(3)
```

```python
# Saving sample master data to csv
updated_master_data.to_csv('Sample_Merged_Dataset.csv', encoding = 'utf-8')
```
