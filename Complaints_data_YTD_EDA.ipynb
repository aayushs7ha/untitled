### Exploratory Data Analysis for Complaints Data

```python
# Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Loading the complaint Data
complaint_ytd = pd.read_excel("Complaints_data_YTD-Big_Data.xltm")
print(complaint_ytd.shape)
complaint_ytd.head(3)
```

```python
# Checking the data types of variables
complaint_ytd.dtypes
```

```python
# Checking for missing values
complaint_ytd.isnull().sum()
```

```python
# Checking the min, max and median values for the numeric variables
print('Min value for Complaint Pending For Days#', complaint_ytd['Complaint Pending For Days#'].min())
print('Max value for Complaint Pending For Days#', complaint_ytd['Complaint Pending For Days#'].max())
print('Median value for Complaint Pending For Days#', complaint_ytd['Complaint Pending For Days#'].median())
print('Average value for Complaint Pending For Days#', complaint_ytd['Complaint Pending For Days#'].mean())
```

```python
# Checking the min, max and median values for the numeric variables
print('Min value for Closing Time', complaint_ytd['Closing Time'].min())
print('Max value for Closing Time', complaint_ytd['Closing Time'].max())
print('Median value for Closing Time', complaint_ytd['Closing Time'].median())
print('Average value for Closing Time', complaint_ytd['Closing Time'].mean())
```

```python
# Checking the min, max and median values for the numeric variables
print('Min value for Mileage', complaint_ytd['Mileage'].min())
print('Max value for Mileage', complaint_ytd['Mileage'].max())
print('Median value for Mileage', complaint_ytd['Mileage'].median())
print('Average value for Mileage', complaint_ytd['Mileage'].mean())
```

```python
# Checking the min, max and median values for the numeric variables
print('Min value for Reopen Count', complaint_ytd['Reopen Count'].min())
print('Max value for Reopen Count', complaint_ytd['Reopen Count'].max())
print('Median value for Reopen Count', complaint_ytd['Reopen Count'].median())
print('Average value for Reopen Count', complaint_ytd['Reopen Count'].mean())
```

```python
# Checking the min, max and median values for the numeric variables
print('Min value for Repeated Count', complaint_ytd['Repeated Count'].min())
print('Max value for Repeated Count', complaint_ytd['Repeated Count'].max())
print('Median value for Repeated Count', complaint_ytd['Repeated Count'].median())
print('Average value for Repeated Count', complaint_ytd['Repeated Count'].mean())
```

```python
# Adding a column for Vehicle Age
complaint_ytd['Vehicle_age'] = 2021 - complaint_ytd['Model Year']
print(complaint_ytd.shape)
complaint_ytd.head(3)
```

```python
# Checking the values in Vehicle age column
complaint_ytd['Vehicle_age'].value_counts().sort_index(ascending = True)
# Most of the complaints come in new vehicles where vehicle age is 1 2 years.
```

```python
# Creating a column to identify the time taken to close the complaints

# Changing the data type of Opened and Closed date to timestamp
complaint_ytd['Opened Date'] = pd.to_datetime(complaint_ytd['Opened Date'])
complaint_ytd['Closed Date'] = pd.to_datetime(complaint_ytd['Closed Date'])
complaint_ytd.dtypes
```

```python
# Getting the time taken in closing the complaint
complaint_ytd['Complaint_closing_time'] = (complaint_ytd['Closed Date'] - complaint_ytd['Opened Date']).dt.days
print(complaint_ytd.shape)
complaint_ytd.head(3)
```

```python
# Now we have the time required to close the complaints in number of days
# Checking the min max, avg and median value for complaint closing time
print('Min value for Complaint Closing Time', complaint_ytd['Complaint_closing_time'].min())
print('Max value for Complaint Closing Time', complaint_ytd['Complaint_closing_time'].max())
print('Median value forComplaint Closing Time', complaint_ytd['Complaint_closing_time'].median())
print('Average value for Complaint Closing Time', complaint_ytd['Complaint_closing_time'].mean())
```
