```python
import os
import re
import shutil as sh

#Appedning content to docx file
def write_append_docx(input_file=None, complete_file_name=None,heading_name=None):
    import docx
    import os
    doc = docx.Document(input_file)  # Creating word reader object.
    data = ""
    fullText = []
    fullText.append(f"**************************************{heading_name}***********************************")
    for para in doc.paragraphs:
        fullText.append(para.text)
        data = '\n'.join(fullText)
    fullText.append(f"*********************************End Of {heading_name}***********************************")
    data = '\n'.join(fullText)

    filename = complete_file_name
    if not (os.path.exists(complete_file_name)):
        document = docx.Document()
        document.save(filename)
    document = docx.Document(filename)
    paragraph = document.add_paragraph(data)
    document.save(filename)
#----------------------------------------------------------------------------------------#
def move_merge_func(source_dir=None, target_dir=None,delete_flag='N'):
    if source_dir==None or target_dir==None:
        print(r'''Example :-- move_merge_func(source_dir=r"C:\Users\aditya\Downloads",
                                             target_dir=r"D:\inueron\Python\examples"),
                                             delete_flag='Y/N'
                                             
                                             ''')
        return None
    filenames = os.listdir(f"{source_dir}")
    new_dir_lst=[]
    for i in filenames:
        dir_name = ''
        if re.match("(Programming_)*Assi.*.docx", i) is not None:
            
            for char in i.split(".")[0]:
                if char.isalpha() or char in "_":
                    dir_name = dir_name + char
                    
            if dir_name[-1] == "_":
                dir_name = dir_name[:-1]
                new_dir_lst.append(dir_name)
            else:
                new_dir_lst.append(dir_name)
                
            if os.path.isdir(f"{target_dir}/{dir_name}"):
                pass

            else:
                os.mkdir(f"{target_dir}/{dir_name}")
                
            ### moving , renaming & appending it file content & did not delete the original file
            if re.match("[() ]*", i) is not None:
                new_filename = re.sub("[() ]*", '', i)
                sh.move(f"{source_dir}/{i}", f"{target_dir}/{dir_name}/{new_filename}")
                print(f"New file name is :- {target_dir}/{dir_name}/{new_filename}")
                write_append_docx(f"{target_dir}/{dir_name}/{new_filename}",
                                  f"{target_dir}/{dir_name}/Complete_{dir_name}.docx",
                                  new_filename.split(".")[0])
            else:
                sh.move(f"{source_dir}/{i}", f"{target_dir}/{dir_name}/{i}")
                print(f"File name is same :- {target_dir}/{dir_name}/{i}")
                write_append_docx(f"{target_dir}/{dir_name}/{i}",
                                  f"{target_dir}/{dir_name}/Complete_{dir_name}.docx",
                                  i.split(".")[0])
        ##delete the original file from the new directory use has the choice 
    if delete_flag=='Y':
        new_dir_lst=set(new_dir_lst)
        print(new_dir_lst)
        for i in new_dir_lst:
            filenames=os.listdir(f"{target_dir}/{i}")
            for j in filenames:
                if 'Complete' not in j:
                    os.remove(f"{target_dir}/{i}/{j}")
                    
move_merge_func()
```
