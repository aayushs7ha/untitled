6.	Uniqueness(duplicacy check )

```python
All=pd.DataFrame()
for df in list_tables_db:
    table_name=eval(df)
    #Profile.to_file(output_file="D:/HMM_DQA/DQA_Results/"+df+".html")
    missing=table_name.isnull().sum()
    missing=pd.DataFrame(missing)
    missing.columns =['Missing column_value']
    missing['Table_Name']=df
    missing['Missing column_value']= missing['Missing column_value'].astype(str)
    missing['Column_Name']=list(table_name.columns)
    missing['Missing_Count'] = missing['Missing column_value'].str.replace('([A-Za-z]+)', '')
    count_row = table_name.shape[0]
    duplicate=table_name.duplicated().sum()
    missing['Total Records']=count_row
    missing['Missing Percent']=missing['Missing_Count'].astype(int)*100/missing['Total Records']
    missing['Duplicate count']=duplicate
    missing['Duplicate Percent']=missing['Duplicate count'].astype(int)*100/missing['Total Records']
    #missing.to_excel("D:/HMM_DQA/DQA_Results/DQA_Metrics/"+df+"_Details.xlsx",sheet_name=df)
    All=All.append(missing)
#All.to_excel("D:/HMM_DQA/DQA_Results/DQA_Metrics/All_Tables_Details.xlsx",sheet_name='All')
```
