```python
# average sold price for each age category
df.groupby('AGE')['SOLD PRICE'].mean()
```

```python
sold_price_by_age = df.groupby('AGE')['SOLD PRICE'].mean().reset_index()
```

```python
sold_price_by_age
```

```python
soldprice_by_age_role = df.groupby(['AGE', 'PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()

soldprice_by_age_role
```

```python
soldprice_comparison = soldprice_by_age_role.merge(sold_price_by_age,on = 'AGE',how = 'outer')
soldprice_comparison
```

```python
# Renaiming coloumns

soldprice_comparison.rename(columns={'SOLD PRICE_x':"SOLD_PRICE_AGE_ROLE", 'SOLD PRICE_y': "SOLD_PRICE_AGE"}, inplace = True)
```

```python
soldprice_comparison.head(5)
```

```python
soldprice_comparison['change'] = soldprice_comparison.apply(lambda rec:
(rec.SOLD_PRICE_AGE_ROLE - rec.SOLD_PRICE_AGE) / rec.SOLD_PRICE_AGE, axis = 1)
```
